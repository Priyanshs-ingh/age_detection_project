```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'merged_data.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Selecting a subset of metrics for correlation
selected_columns = ['1990', '2021.0', 'Absolute Change', 'Relative Change', 'Smoking_2021', 'Population']
correlation = data[selected_columns].corr()

# Creating a more aesthetic heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='Blues', fmt=".2f", linewidths=.5, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap of Key Health Metrics')
plt.show()

```

```python
# Pair Plot for selected variables
sns.pairplot(data[selected_columns], diag_kind='kde')
plt.suptitle('Pairwise Relationships of Key Metrics', size=16, y=1.02)
plt.show()

# Box Plots for Smoking Rates and a Disease metric
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.boxplot(y=data['Smoking_2021'], ax=axes[0], color='skyblue')
axes[0].set_title('Box Plot of Smoking Rates in 2021')


plt.tight_layout()
plt.show()

# Density Plots for comparisons over years
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(data['1990'], label='1990', shade=True, color='steelblue')
sns.kdeplot(data['2021.0'], label='2021', shade=True, color='darkred')
plt.title('Density Plot of Health Metrics Across Years')
plt.legend()
plt.show()
```

Two Peaks (Modes): The density plot shows two peaks or modes for each year, suggesting that the distribution of the health metric might be bimodal in both years. This could indicate two distinct groups or types within the data that differ significantly in their health metric values.

Shift in Distribution:

1990 (Blue Curve): The peak is sharper and centered around a lower value (approximately between 0 and 10). This suggests that in 1990, a larger proportion of the dataset had lower health metric values, and the values were more concentrated around this peak.
2021 (Red Curve): The peak is broader and shifted towards higher values (centered around 10 to 20), indicating an overall increase in the health metric values over time. The broader peak also suggests increased variability in the health metric in 2021 compared to 1990.
Change in Density:

The area under each curve represents the probability distribution of the health metrics in each year. The red curve’s broader spread indicates more diverse health metric outcomes in 2021 than in 1990, which had a more concentrated distribution of values.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load your dataset
data_path = 'updated_final_merged_data1.csv'
data = pd.read_csv(data_path)

# Setup the plot with a larger figure size
fig, ax = plt.subplots(figsize=(18, 10))  # Increased figure size

# Assuming the data columns are named as described above
factors = ['Total Change', 'Due to Ageing', 'Due to Prevalence Changes', 'Due to Population Growth']
colors = {'Total Change': 'black', 'Due to Ageing': 'orange', 'Due to Prevalence Changes': 'green', 'Due to Population Growth': 'blue'}
markers = {'Female': 'o', 'Male': 's'}

# Define a wider range for x-positions to avoid overlapping
region_count = len(data['Region'].unique())
x_positions = np.linspace(0, region_count * 2, region_count)  # More space between each region

# Iterate through each type (Current and Former)
for i, smoker_type in enumerate(['Current', 'Former']):
    group_data = data[data['Type'] == smoker_type]

    for gender in ['Female', 'Male']:
        gender_data = group_data[group_data['Gender'] == gender]
        offset = 0.4 if gender == 'Male' else -0.4  # Increased offset
        
        for factor in factors:
            # Apply offset to x_positions based on gender
            adjusted_x_positions = x_positions + offset
            ax.scatter(adjusted_x_positions, gender_data[factor], color=colors[factor], marker=markers[gender], label=f'{factor} ({gender})' if i == 0 and gender == 'Female' else "")

# Set labels and titles
ax.set_xticks(x_positions)  # Set x-ticks to the middle of the groups
ax.set_xticklabels(data['Region'].unique(), rotation=45, ha='right',fontsize=6)
ax.set_ylabel('Percentage Change (%)')
ax.set_title('Percentage Change in Smoking Prevalence by Gender and Factor')
ax.grid(True)

# Create a legend
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))  # Remove duplicates
ax.legend(by_label.values(), by_label.keys(), title='Legend', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(0)

# Generate years
years = np.arange(1990, 2051)

# Generate base, decline, and stable data for morbidity and rate
# Base levels before 2023
base_morbidity_female = 800 + np.random.normal(scale=30, size=years[years < 2023].size)
base_morbidity_male = 800 + np.random.normal(scale=30, size=years[years < 2023].size)
base_rate_female = 30000 + np.random.normal(scale=1000, size=years[years < 2023].size)
base_rate_male = 30000 + np.random.normal(scale=1000, size=years[years < 2023].size)

# Declining phase from 2023 to 2050
decline_morbidity_female = np.linspace(800, 100, num=years[(years >= 2023) & (years <= 2050)].size) + np.random.normal(scale=10, size=years[(years >= 2023) & (years <= 2050)].size)
decline_morbidity_male = np.linspace(800, 100, num=years[(years >= 2023) & (years <= 2050)].size) + np.random.normal(scale=10, size=years[(years >= 2023) & (years <= 2050)].size)
decline_rate_female = np.linspace(30000, 5000, num=years[(years >= 2023) & (years <= 2050)].size) + np.random.normal(scale=500, size=years[(years >= 2023) & (years <= 2050)].size)
decline_rate_male = np.linspace(30000, 5000, num=years[(years >= 2023) & (years <= 2050)].size) + np.random.normal(scale=500, size=years[(years >= 2023) & (years <= 2050)].size)

# Level off after 2050
stable_morbidity_female = np.full(years[years > 2050].size, 100 + np.random.normal(scale=5, size=years[years > 2050].size))
stable_morbidity_male = np.full(years[years > 2050].size, 100 + np.random.normal(scale=5, size=years[years > 2050].size))
stable_rate_female = np.full(years[years > 2050].size, 5000 + np.random.normal(scale=250, size=years[years > 2050].size))
stable_rate_male = np.full(years[years > 2050].size, 5000 + np.random.normal(scale=250, size=years[years > 2050].size))

# Combine the segments
morbidity_female = np.concatenate([base_morbidity_female, decline_morbidity_female, stable_morbidity_female])
morbidity_male = np.concatenate([base_morbidity_male, decline_morbidity_male, stable_morbidity_male])
rate_female = np.concatenate([base_rate_female, decline_rate_female, stable_rate_female])
rate_male = np.concatenate([base_rate_male, decline_rate_male, stable_rate_male])

# Create DataFrame
data = pd.DataFrame({
    'Year': years,
    'Female Morbidity': morbidity_female,
    'Male Morbidity': morbidity_male,
    'Female Rate': rate_female,
    'Male Rate': rate_male
})

# Plotting the data
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10), sharex='col', sharey='row')

# Set the vertical reference lines
for ax in axs.flat:
    ax.axvline(x=2023, color='black', linestyle='--', label='Reference-2023')
    ax.axvline(x=2050, color='red', linestyle='--', label='Elimination-2050')
    ax.legend()

# Plotting Morbidity and Rate for Female and Male
axs[0, 0].plot(data['Year'], data['Female Morbidity'], color='blue')
axs[0, 0].fill_between(data['Year'], data['Female Morbidity'] - 50, data['Female Morbidity'] + 50, color='gray', alpha=0.5)
axs[0, 0].set_title('Female Morbidity')
axs[0, 0].set_ylabel('Morbidity (millions)')

axs[0, 1].plot(data['Year'], data['Male Morbidity'], color='blue')
axs[0, 1].fill_between(data['Year'], data['Male Morbidity'] - 50, data['Male Morbidity'] + 50, color='gray', alpha=0.5)
axs[0, 1].set_title('Male Morbidity')

axs[1, 0].plot(data['Year'], data['Female Rate'], color='blue')
axs[1, 0].fill_between(data['Year'], data['Female Rate'] - 1000, data['Female Rate'] + 1000, color='gray', alpha=0.5)
axs[1, 0].set_title('Female Rate per 100,000')
axs[1, 0].set_ylabel('Rate per 100,000')

axs[1, 1].plot(data['Year'], data['Male Rate'], color='blue')
axs[1, 1].fill_between(data['Year'], data['Male Rate'] - 1000, data['Male Rate'] + 1000, color='gray', alpha=0.5)
axs[1, 1].set_title('Male Rate per 100,000')

plt.tight_layout()
plt.show()

# Plotting the revised data
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 10), sharex='col', sharey='row')

# Setting common styles
for ax in axs.flat:
    ax.axvline(x=2023, color='black', linestyle='--', label='Reference-2023')
    ax.axvline(x=2050, color='red', linestyle='--', label='Elimination-2050')
    ax.legend()

# Morbidity plots
axs[0, 0].plot(years, morbidity_female, color='blue')
axs[0, 0].fill_between(years, morbidity_female - 50, morbidity_female + 50, color='gray', alpha=0.5)
axs[0, 0].set_title('Female Morbidity')
axs[0, 0].set_ylabel('Morbidity (millions)')

axs[0, 1].plot(years, morbidity_male, color='blue')
axs[0, 1].fill_between(years, morbidity_male - 50, morbidity_male + 50, color='gray', alpha=0.5)
axs[0, 1].set_title('Male Morbidity')

# Rate plots
axs[1, 0].plot(years, rate_female, color='blue')
axs[1, 0].fill_between(years, rate_female - 1000, rate_female + 1000, color='gray', alpha=0.5)
axs[1, 0].set_title('Female Rate per 100,000')
axs[1, 0].set_ylabel('Rate per 100,000')

axs[1, 1].plot(years, rate_male, color='blue')
axs[1, 1].fill_between(years, rate_male - 1000, rate_male + 1000, color='gray', alpha=0.5)
axs[1, 1].set_title('Male Rate per 100,000')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

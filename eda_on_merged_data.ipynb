```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = 'merged_data.csv'  # Replace with the actual path to your CSV file
data = pd.read_csv(file_path)

# Selecting a subset of metrics for correlation
selected_columns = ['1990', '2021.0', 'Absolute Change', 'Relative Change', 'Smoking_2021', 'Disease1', 'Population']
correlation = data[selected_columns].corr()

# Creating a more aesthetic heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='Blues', fmt=".2f", linewidths=.5, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap of Key Health Metrics')
plt.show()

```

```python
# Pair Plot for selected variables
sns.pairplot(data[selected_columns], diag_kind='kde')
plt.suptitle('Pairwise Relationships of Key Metrics', size=16, y=1.02)
plt.show()

# Box Plots for Smoking Rates and a Disease metric
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.boxplot(y=data['Smoking_2021'], ax=axes[0], color='skyblue')
axes[0].set_title('Box Plot of Smoking Rates in 2021')
sns.boxplot(y=data['Disease1'], ax=axes[1], color='lightgreen')
axes[1].set_title('Box Plot of Disease1 Prevalence')
plt.tight_layout()
plt.show()

# Density Plots for comparisons over years
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(data['1990'], label='1990', shade=True, color='steelblue')
sns.kdeplot(data['2021.0'], label='2021', shade=True, color='darkred')
plt.title('Density Plot of Health Metrics Across Years')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data
file_path = 'merged_data.csv'  # Make sure to replace this with your file path
data = pd.read_csv(file_path)

# Enhanced Pair Plot with regression
sns.pairplot(data[['1990', '2021.0', 'Smoking_2021', 'Disease1']], kind='reg', diag_kind='kde')
plt.suptitle('Pair Plot with Regression of Key Metrics')
plt.show()

# Box Plots for Key Metrics
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=data['Smoking_2021'], color='lightblue')
plt.title('Box Plot of Smoking Rates in 2021')

plt.subplot(1, 2, 2)
sns.boxplot(y=data['Disease1'], color='lightgreen')
plt.title('Box Plot of Disease1 Prevalence')
plt.tight_layout()
plt.show()

# Time Trend Comparison
plt.figure(figsize=(10, 5))
plt.plot(data['1990'], label='1990 Health Metric')
plt.plot(data['2021.0'], label='2021 Health Metric', color='red')
plt.title('Health Metric Comparison Over Time')
plt.legend()
plt.show()

# Density Plots for Metric Distribution
plt.figure(figsize=(10, 5))
sns.kdeplot(data['1990'], label='1990', shade=True)
sns.kdeplot(data['2021.0'], label='2021', shade=True, color='red')
plt.title('Density Plot of Health Metrics Across Years')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, y='Smoking_2021', inner='quartile', palette='light:g')
plt.title('Violin Plot of Smoking Rates in 2021')
plt.show()

```

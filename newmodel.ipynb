```python
import pandas as pd

# Load the cleaned data
data = pd.read_csv('cleaned_age_by_category.csv')

# Check for non-numeric columns and convert them
for col in data.columns:
    if data[col].dtype == 'object':
        try:
            data[col] = pd.to_numeric(data[col], errors='raise')
        except ValueError:
            print(f"Cannot convert column {col} to numeric.")

# Now check the datatypes again
print(data.dtypes)
```

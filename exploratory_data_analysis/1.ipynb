```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulating df_smoking_rates
df_smoking_rates = pd.DataFrame({
    'year': [2015, 2016, 2017, 2018, 2019],
    'Heavy smokers': [200, 190, 180, 170, 160],
    'Moderate smokers': [300, 310, 320, 330, 340],
    'Light smokers': [500, 490, 480, 470, 460]
})

# Simulating df_health_outcomes (assuming death counts are related to smoking)
df_health_outcomes = pd.DataFrame({
    'year': [2015, 2016, 2017, 2018, 2019],
    'Deaths from heavy smoking': [50, 48, 46, 44, 42],
    'Deaths from moderate smoking': [30, 31, 32, 33, 34],
    'Deaths from light smoking': [10, 9, 8, 7, 6]
})

# Merging datasets on 'year'
merged_data = pd.merge(df_smoking_rates, df_health_outcomes, on='year')

# Creating plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting the number of smokers in each category
ax1.bar(merged_data['year'] - 0.2, merged_data['Heavy smokers'], width=0.2, color='darkred', label='Heavy Smokers')
ax1.bar(merged_data['year'], merged_data['Moderate smokers'], width=0.2, color='orange', label='Moderate Smokers')
ax1.bar(merged_data['year'] + 0.2, merged_data['Light smokers'], width=0.2, color='yellow', label='Light Smokers')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Smokers')
ax1.legend(loc='upper left')

# Adding a secondary axis for deaths
ax2 = ax1.twinx()
ax2.plot(merged_data['year'], merged_data['Deaths from heavy smoking'], 'r-o', label='Deaths - Heavy Smoking')
ax2.plot(merged_data['year'], merged_data['Deaths from moderate smoking'], 'b-o', label='Deaths - Moderate Smoking')
ax2.plot(merged_data['year'], merged_data['Deaths from light smoking'], 'g-o', label='Deaths - Light Smoking')
ax2.set_ylabel('Number of Deaths')
ax2.legend(loc='upper right')

plt.title('Smoking Habits vs. Smoking Related Deaths Over the Years in USA ')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv(r'C:\Users\priyansh singh\Documents\GitHub\age_detection_project\Current Cigarette Use Among Adults (Behavior Risk Factor Surveillance System) 2018.csv')

# Define thresholds for categorization (adjust these based on your data analysis)
low_threshold = 15
moderate_threshold_low = 15
moderate_threshold_high = 30
high_threshold = 30

# Define conditions and categories for smoker categorization based on confidence limits and Data_Value
conditions = [
    (df['Low_Confidence_Limit'] > high_threshold),  # High smokers: lower bound is above high threshold
    (df['Data_Value'] > moderate_threshold_low) & (df['Data_Value'] <= moderate_threshold_high),  # Moderate smokers: data value within moderate range
    (df['High_Confidence_Limit'] < low_threshold)  # Low smokers: upper bound is below low threshold
]
categories = ['High', 'Moderate', 'Low']
df['Smoker_Category'] = np.select(conditions, categories, default='Uncategorized')

# Display categorized data
print(df[['Location', 'Data_Value', 'Low_Confidence_Limit', 'High_Confidence_Limit', 'Smoker_Category']].head())

# Group by location and category to summarize data
grouped_df = df.groupby(['Location', 'Smoker_Category']).size().unstack(fill_value=0)
print(grouped_df)

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
grouped_df.plot(kind='bar', stacked=True, ax=ax)
plt.title('Number of Smokers by Category and Location')
plt.xlabel('Location')
plt.ylabel('Number of Smokers')
plt.xticks(rotation=45)
plt.legend(title='Smoker Category')
plt.tight_layout()
plt.show()
```

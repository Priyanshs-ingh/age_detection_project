```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
df_smoking_rates = pd.read_csv(r'C:\Users\priyansh singh\Documents\GitHub\age_detection_project\table_e90dbf0b-10da-42c9-9938-f13a24c38101.csv')
df_cigarette_use_2018 = pd.read_csv(r'C:\Users\priyansh singh\Documents\GitHub\age_detection_project\Current Cigarette Use Among Adults (Behavior Risk Factor Surveillance System) 2018.csv')

# Add a 'year' column to df_cigarette_use_2018
df_cigarette_use_2018['year'] = 2018

# Perform the merge
merged_df = pd.merge(df_smoking_rates, df_cigarette_use_2018, on='year', how='inner')

# Example plotting (adjust column names based on your merged dataframe)
plt.figure(figsize=(10, 5))
plt.plot(merged_df['year'], merged_df['Heavy smokers'], label='Heavy Smokers')
# Adjust these plot lines based on the actual columns in merged_df
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Smoking Rates Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df_death_causes = pd.read_csv(r'C:\Users\priyansh singh\Documents\GitHub\age_detection_project\table_1fdb7992-7c9a-4fb3-9504-f8cce7e0154d.csv')

# Summing up all causes for a simple visualization (assumes each row might be a different category if available)
death_causes_sum = df_death_causes[['Non-natural cause', 'Other natural cause', 'Respiratory disease', 'Cardiovascular disease', 'Other form of cancer', 'Lung cancer']].sum()
death_causes_sum.plot(kind='bar')
plt.title('Sum of Causes of Death from Dataset')
plt.ylabel('Count')
plt.xlabel('Cause of Death')
plt.show()
```

```python
sns.clustermap(df_smoking_rates.drop('year', axis=1), standard_scale=1)
plt.title('Cluster Map of Smoking Rates')
plt.show()
```

```python
pd.plotting.scatter_matrix(df_smoking_rates.drop('year', axis=1), alpha=0.2, figsize=(10, 10))
plt.title('Scatter Plot Matrix of Smoking Data')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(x='Location', y='Data_Value', data=df_cigarette_use_2018)
plt.xticks(rotation=90)
plt.title('Distribution of Smoking Rates by State (2018)')
plt.xlabel('State')
plt.ylabel('Smoking Rate')
plt.show()
```

```python
df_alzheimers_data['Data_Value'].hist(by=df_alzheimers_data['YearStart'])
plt.title('Alzheimer\'s Disease Data Distribution by Year')
plt.xlabel('Data Value')
plt.ylabel('Frequency')
plt.show()
```

```python
for year in df_smoking_rates['year'].unique():
    subset = df_smoking_rates[df_smoking_rates['year'] == year]
    sns.kdeplot(subset['Heavy smokers'], label=str(year))
plt.title('Density Plot of Heavy Smokers Over Years')
plt.xlabel('Heavy Smokers Percentage')
plt.legend(title='Year')
plt.show()
```
